```python
from decimal import Decimal
import pandas as pd
from IPython.display import display


from IPython.display import display_html
def display_side_by_side(*args):
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline; marging-left: 1em"'),raw=True)

########## Parametros
#Excluir essas palavras
EXCLUDE_WORDS = '/bparcela|parcelas|vezes|parcelada|parcelado|entrada|financiamento/b'

# PREÇO MAXIMO E MINIMO
MAX_PRICE = 30000
MIN_PRICE = 10000

df = pd.read_csv('result.csv')
df.head()
```

```python
########### Cleaning
df.dropna(axis="rows",subset=['price'])

df.price = df.price.map(lambda x: float(str(x).replace("R$", '').replace('.','').strip()) )
df.description = df.description.map(lambda x: x.strip())
df = df[~df['description'].str.contains(EXCLUDE_WORDS, case=False, regex=True)]


##########FILTERING
df = df[df['price'] <= MAX_PRICE]
df = df[df['price'] >= MIN_PRICE]
```

```python
# AVARAGE PRICE
grouped_df = df.groupby(["Município"])
DESVIO_PADRAO = grouped_df['price'].std(ddof=0).round(2)
DESVIO_PADRAO = pd.DataFrame(DESVIO_PADRAO).reset_index()
DESVIO_PADRAO.columns = ['Município', 'desviopd_municipio']
                             
grouped_df = df.groupby(["Município"])
MEAN = grouped_df['price'].mean().round(2)
MEAN = pd.DataFrame(MEAN).reset_index()
MEAN.columns = ['Município', 'media_municipio']

grouped_df = df.groupby(["Município"])
VARIANCIA = grouped_df['price'].var(ddof=0).round(2)
VARIANCIA = pd.DataFrame(VARIANCIA).reset_index()
VARIANCIA.columns = ['Município', 'variancia']


df = pd.merge(df, DESVIO_PADRAO,on='Município')
df = pd.merge(df, MEAN,on='Município')
df = pd.merge(df, VARIANCIA,on='Município')

#DESVIO_PADRAO.head(50)
display_side_by_side(MEAN,DESVIO_PADRAO, VARIANCIA)
#pd.DataFrame(grouped_df['price'])
```

```python
######### SORTING
df = df.sort_values(by=['desviopd_municipio'], ascending=False)
# Observe o valor do terreno em relação a média e observe o desvio padrão. Terrenos com preço muito abaixo da média são possíveis bons negócios para a revenda. Estes terrenos devem ser analisados com mais atenção observando a infraestrutura da região.
pd.set_option('display.max_rows', df.shape[0]+1)
df
```
